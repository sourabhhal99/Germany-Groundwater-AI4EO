```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('germany_water.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
plt.plot(df['Gravity'],color = 'black')
plt.title('Gravity Graph')
plt.xlabel('Ground Water')
#plt.ylabel('Date')
plt.grid(True)
plt.show()


plt.figure(figsize=(4,4))
plt.plot(df['Rainfall'],color = 'red')
plt.title('Rainfall Graph')
plt.xlabel('Rainfall')
#plt.ylabel('Date')
plt.grid(True)
plt.show()


plt.figure(figsize=(4,4))
plt.plot(df['Temperature'],color = 'red')
plt.title('Temprature Graph')
plt.xlabel('Temprature')
#plt.ylabel('Date')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sets Graph size
fig, ax1 = plt.subplots(figsize=(12, 5))

# --- 1st LINE: GRAVITY (Black) ---
# ax1 means Left scale
ax1.plot(df['Gravity'], color='black', label='Gravity (Paani)')
ax1.set_ylabel('Groundwater Level', color='black') 
ax1.tick_params(axis='y', labelcolor='black')      

# --- 2nd LINE: RAINFALL (Blue) ---
# ax2 means Right scale (twinx axis)
ax2 = ax1.twinx()
ax2.plot(df['Rainfall'], color='blue', linestyle='--', label='Rainfall')
ax2.set_ylabel('Rainfall Amount', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Comparison 1: Rainfall vs Groundwater')
plt.show()
```
